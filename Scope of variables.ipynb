The part of program where a variable is accessible is called it's scope
Scope is the visibility of a variable declared within a specific region

* there are two types of scopes
Local scope 
Global scope

# Local Scope

```python
# local Scope 
def func():
    a = 20
    print(a)
    
```

```python
func()
```

```python
a
```

* Variable created inside a scope is a Local scope

 # Global Scope

* Variable created inside the main body of the python code is Global scope

```python
# Global Scope
x = 10

def myfunc():
    y = 7
    print(x)
```

```python
myfunc()
```

```python
# Ninja Trick
def myfunc1():
    global y
    y = 7
    print(y)
```

```python
myfunc1()
```

```python
y
```
